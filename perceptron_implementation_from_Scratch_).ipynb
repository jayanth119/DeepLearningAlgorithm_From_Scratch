```python
from sklearn.datasets import make_classification
import numpy as np
X,Y = make_classification(n_samples = 169 , n_features= 2 , n_redundant= 0, n_informative = 2 , n_classes = 2 , random_state=69, n_clusters_per_class= 1 )
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize = (10, 6))
plt.scatter(X[: , 0 ] , X[: , 1]  ,c = Y ,cmap='winter')
```

```python
def step(x):
  return 1 if x>0 else 0
```

```python
import random as rd
import math as mt
def Perceptron(X, Y, epochs):
  X = np.insert(X,0,1,axis = 1)
  weights = np.ones(X.shape[1])
  learning_rate = 0.1
  for i in range(epochs ):
    p = rd.randint(0,100) #pick random point from data
    y_dash = step(np.dot(X[p],weights)) #check that point classified region
    #formula
    # newweight  = oldweights + (orginalclass - calulatedclass)*(picked point )
    weights = weights + (Y[p] - y_dash)*X[p]
  return weights[0] ,weights[1:]

```

```python
coeff , intercept = Perceptron(X,Y, 1000)
print("coefficent of the points ",intercept )
print("intercept of the points", coeff)
```

**above code process can have some drawbacks to overcome the we introduced loss functions**

```python
def Perceptron_Loss_Function(X,Y , epochs):
  w1 = w2 = b = 1
  learning_rate = 0.1
  for i in range(epochs):
    for  j in range(X.shape[0]):
      z = w1*X[j][0] + w2*X[j][1] + b
      if(z*Y[j]<0):
        w1 = w1 + learning_rate*Y[j]*X[j][0]
        w2 = w2 + learning_rate*Y[j]*X[j][1]
        b  = b + learning_rate*Y[j]
  return w1 , w2, b



```

```python
w1 , w2, b  = Perceptron_Loss_Function(X,Y,1000)
print("coeffient  is " , w1 , 'and ' , w2)
print("intercept is " , b )
```
